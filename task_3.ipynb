```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

#     
x_data = np.arange(0, 11, 1)         # x = 0, 1, 2, ..., 10
y_data = (-np.cos(x_data)) ** 3      # f(x) = (-cos(x))^3

#      
x_dense = np.linspace(0, 10, 100, endpoint=True)

#     
linear_interpolation = interp1d(x_data, y_data, kind='linear')  #  
cubic_interpolation = interp1d(x_data, y_data, kind='cubic')    #  

#  
plt.plot(x_data, y_data, 'o', label=' ')                   #  
plt.plot(x_dense, linear_interpolation(x_dense), '-', label=' ') #  
plt.plot(x_dense, cubic_interpolation(x_dense), '--', label=' ') #  

plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("    f(x) = (-cos(x))^3")
plt.grid(True)
plt.show()
```
