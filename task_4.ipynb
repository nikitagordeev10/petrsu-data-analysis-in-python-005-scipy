```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
```

```python
# Параметры равномерного распределения
a, b = 10, 20  # Интервал [10, 20]

# Генерация случайной выборки из 20 чисел, Нормальное распределение
random_sample = uniform.rvs(loc=a, scale=b - a, size=20)

print("случайная выборка из 20 чисел:", random_sample)
```

```python
# Значение плотности и функции распределения в точке x = 15
x_value = 15
pdf_value = uniform.pdf(x_value, loc=a, scale=b - a)  # Плотность распределения
cdf_value = uniform.cdf(x_value, loc=a, scale=b - a)  # Функция распределения

print("Значения плотности в точке x = 15:", pdf_value)
print("Значения функции распределения в точке x = 15", cdf_value)
```

```python
# Подготовка данных для графиков
x = np.linspace(a, b, 500)
pdf = uniform.pdf(x, loc=a, scale=b - a)  # Плотность распределения
cdf = uniform.cdf(x, loc=a, scale=b - a)  # Функция распределения

# Построение графиков
plt.figure(figsize=(12, 6))

# График плотности распределения
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label="Плотность распределения")
plt.axvline(x=x_value, color='red', linestyle='--', label=f"x = {x_value}")
plt.title("Плотность распределения")
plt.xlabel("x")
plt.ylabel("Плотность")
plt.legend()
plt.grid(True)

# График функции распределения
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label="Функция распределения", color="green")
plt.axvline(x=x_value, color='red', linestyle='--', label=f"x = {x_value}")
plt.title("Функция распределения")
plt.xlabel("x")
plt.ylabel("Вероятность")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
